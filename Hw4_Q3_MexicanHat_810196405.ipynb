```python
# Mexican Hat:
import numpy as np


def activation_Func(X1,x_max):
    X=X1
    for i in range(X.shape[0]):
         if (X[i]<0):
             X[i]=0
         elif(X[i]>x_max):
             X[i]=x_max
    return X         


X=np.asarray([0.32,0.33,0.28,0.47,0.66,0.80,0.4,0.33,0.1,0.26])



# Algo: Step1:
R1=1
R2=5
t_max=5

x_max=2 # Given in Question 3
C1=np.random.rand(1)
C2=-np.random.rand(1)
X_old=0
# Step2: t=0
X_old=X
X_in_Iters=[] # to Save X_old in each state (t)
for t in range(1,t_max+1):
    Prev=np.zeros(X.shape[0])
    for i in range(X.shape[0]):
        for k in range(-R1,R1+1):
            if(i+k in range(X.shape[0])): # Step3 : Compute Net input
                Prev[i]=Prev[i]+C1*X_old[i+k] # First Term --> Update
        for k in range(-R2,-R1):
            if(i+k in range(X.shape[0])):
                Prev[i]=Prev[i]+C2*X_old[i+k] # Second Term       
        for k in range(R1+1,R2+1):
            if(i+k in range(X.shape[0])):
                Prev[i]=Prev[i]+C2*X_old[i+k]# Third Term
    # Apply Activation Function: Step 4
    Active_Out=activation_Func(Prev,x_max)
    X_old=Active_Out
    X_in_Iters.append(X_old)
    # Step6: increment t by +1. (used in for loop )


```

```python
import matplotlib.pyplot as plt
Sample=range(len(X))
plt.figure()
plt.plot(Sample, X)
legend = []
for i in range(t_max):
    plt.plot(Sample,X_in_Iters[i])
for i in range(t_max+1):
    legend.append("t= " +str(i))

plt.title('  ** OutPUT for Each State **  ')
plt.grid()
plt.legend(legend)
plt.show()
```
